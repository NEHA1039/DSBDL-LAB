```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=df[['SepalLength','SepalWidth','PetalLength','PetalWidth']].dropna()
y=df['Species'].loc[x.index]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(
    x,y,test_size=0.2,random_state=0, stratify=y
)
```

```python
from sklearn .preprocessing import StandardScaler
sc=StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.transform(xtest)
```

```python
from sklearn.naive_bayes import GaussianNB
guassian = GaussianNB()
guassian.fit(xtrain, ytrain)
```

```python
y_pred=guassian.predict(xtest)
```

```python
from sklearn.metrics import precision_score,confusion_matrix,accuracy_score,recall_score
cm= confusion_matrix(ytest, y_pred)
accuracy_val = accuracy_score(ytest,y_pred)
error_rate=1-accuracy_val
print("Accuracy:", accuracy_val)
```

```python
print("Precision:",precision_score(ytest,y_pred, average='weighted'))
```

```python
print("Recall:",recall_score(ytest,y_pred, average='weighted'))
```

```python
print("Error rate:",error_rate)
```

```python
print("Confusion matrix:\n",cm)
```
